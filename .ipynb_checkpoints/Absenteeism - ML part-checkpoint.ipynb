```python
import pandas as pd
import numpy as np
```

```python
data_preprocessed = pd.read_csv('Absenteeism_preprocessed.csv')
```

## Create the targets

```python
data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours']>data_preprocessed['Absenteeism Time in Hours'].median(),1,0)
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets

data_preprocessed
```

```python
data_preprocessed = data_preprocessed.drop(['Absenteeism Time in Hours', 'Day of the week','Daily Work Load Average', 'Distance to Work'],axis=1)
data_preprocessed.head()
```

```python
data_with_targets = data_preprocessed.copy()
```

## Select the inputs for regression

```python
data_with_targets.iloc[:,:-1]
```

```python
unscaled_inputs = data_with_targets.iloc[:,:-1]
unscaled_inputs.columns.values
```

## Standardize the data

```python
#from sklearn.preprocessing import StandardScaler
```

```python
#absenteeism_scaler = StandardScaler()

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator, TransformerMixin):
    
    def __init__ (self,columns):
        self.scaler = StandardScaler()
        self.columns=columns
        self.mean_=None
        self.var_= None
        
    def fit(self, X, y =None):
        self.scaler.fit(X[self.columns],y)
        self.mean_= np.mean(X[self.columns])
        self.var_=np.var(X[self.columns])
        return self
    def transform (self, X, y=None, copy=None):
        init_col_order = X.columns 
        X_scaled=pd.DataFrame(self.scaler.transform(X[self.columns]),columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled,X_scaled], axis=1)[init_col_order]






```

```python
columns_to_omit = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Education']

columns_to_scale = [x for x in unscaled_inputs.columns.values if x not in columns_to_omit]

absenteeism_scaler = CustomScaler(columns_to_scale)
```

```python

absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
scaled_inputs
```

## Shuffle and split the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size=0.8, random_state=20)
```

## Logistic regression with sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

## Manually check the accuracy

```python
model_outputs = reg.predict(x_train)

np.sum(model_outputs == y_train)/model_outputs.shape[0]
```

```python
feature_name = unscaled_inputs.columns.values

summary_table = pd.DataFrame(columns=['Feature_name'], data=feature_name)
```

```python
summary_table['Coefficient'] = np.transpose(reg.coef_)

summary_table.index = summary_table.index+1
```

```python
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
```

```python
summary_table 

```

```python
summary_table = summary_table.sort_index()
```

```python
summary_table
```

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
summary_table
```

## Testing the model

```python
reg.score(x_test,y_test)
```
